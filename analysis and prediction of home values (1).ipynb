```python
import pandas as pd
import sagemaker

s3 = boto3.client('s3')
local_file_path = '/Users/fhariyaaseem/Downloads/Metro_zhvi_uc_sfrcondo_tier_0.33_0.67_sm_sa_month.csv'
s3.download_file(bucket_name, key,local_file_path)

df = pd.read_csv(local_file_path)

print(df.columns)
```

```python
import pandas as pd

df_long = df.melt(id_vars=['RegionID', 'SizeRank', 'RegionName', 'RegionType', 'StateName'],
                  var_name='Date',
                  value_name='ZHVI')

df_long['Date'] = pd.to_datetime(df_long['Date'])
df_long['ZHVI'].fillna(method='ffill', inplace=True)

print(df_long.head())
```

```python
df_long.to_csv('preprocessed_data.csv', index=False)
```

```python
from IPython.display import FileLink
display(FileLink('preprocessed_data.csv'))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
sns.lineplot(data=df_long, x='Date', y='ZHVI', ci=None)
plt.title('ZHVI Trends Over Time')
plt.xlabel('Date')  # Label for the x-axis
plt.ylabel('ZHVI')  # Label for the y-axis
plt.show()


```

```python
plt.figure(figsize=(14, 8))
sns.lineplot(data=df_long, x='Date', y='ZHVI', hue='StateName', ci=None)
plt.title('ZHVI Trends Over Time by State')
plt.xlabel('Date')
plt.ylabel('ZHVI')
plt.legend(title='State', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=df_long, x='StateName', y='ZHVI')
plt.xticks(rotation=45)
plt.title('Distribution of ZHVI Across Different States')
plt.xlabel('State')
plt.ylabel('ZHVI')
plt.show()
```

```python
df_long['MoM_ZHVI_Change'] = df_long.groupby(['RegionName'])['ZHVI'].pct_change() * 100
plt.figure(figsize=(14, 8))
sns.lineplot(data=df_long, x='Date', y='MoM_ZHVI_Change', hue='RegionName', ci=None)
plt.title('Month-over-Month Percentage Change in ZHVI')
plt.xlabel('Date')
plt.ylabel('MoM Change (%)')
plt.legend(title='Region', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.axhline(0, color='black', linewidth=0.8, linestyle='--')
plt.show()
```
